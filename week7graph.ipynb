```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()

# Add edges based on the updated network after removing 'Abhijit'
G.add_edges_from([
    ('Fiona', 'Souvik'),
    ('Fiona', 'Srinivas'),
    ('Srinivas', 'George'),
    ('George', 'Liam'),
    ('Ramesh', 'Liam'),
    ('Hui', 'Json'),
    ('Json', 'Mike')
])

# Draw the graph
plt.figure(figsize=(10,7))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=12, font_weight='bold', edge_color='gray')
plt.title("Updated Social Network After Removing 'Abhijit'")
plt.show()

# Calculate Degree and Closeness Centrality
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)

# Convert to sorted list to see top nodes
sorted_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
sorted_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)

print("Top Nodes by Degree Centrality:")
for node, centrality in sorted_degree:
    print(f"{node}: {centrality:.2f}")

print("\nTop Nodes by Closeness Centrality:")
for node, centrality in sorted_closeness:
    print(f"{node}: {centrality:.2f}")
```
